```python
import numpy as np
import matplotlib.pyplot as plt

def post_process(data):
    import numpy as np
    
    # 创建一个示例数据集
    data = np.array(data)

    # 计算平均数
    mean_value = np.mean(data)

    # 计算方差
    variance_value = np.var(data)

    return round(mean_value, 2), round(variance_value, 2)

# result_dict {weight : results}
result_dict = {0 : [55.67,55.8],
               20  : [54.17, 54.26, 55.16, 55.63,56.21],
               40  : [53.93, 54.6, 54.82, 54.86,55.24],
               60  : [54.18, 54.93, 55, 55.04, 55.74],
               80  : [54.67, 54.75, 55.06, 55.26, 56.3],
               100 : [54.58, 55.18, 54.31, 54.93, 54.05, 53.69],
               120 : [54.27, 54.64, 54.65, 54.19, 53.79, 53.94],
               140 : [53.45, 53.85, 54.15, 54.77, 54.26],
               160 : [53.88, 54.01, 54.42, 54.6,54.83 ],
               180 : [54.04, 54.33, 53.63, 54.97],
               200 : [54.74, 55.22]}

for key, value in result_dict.items():
    mean, std = post_process(value)
    print(f'weight: {key}: {mean} +- {std}')
    
title = 'RMSE(mm) Under Different Weight of Reconsturction moduel'
x_lable = 'RMSE(mm)'
y_lable = 'R'
weights = list(result_dict.keys())
mean_values = [round(np.mean(result_dict[weight]),2) for weight in weights]
std_values = [round(np.std(result_dict[weight]),2) for weight in weights]
```

```python

box_data = [result_dict[weight] for weight in weights]


# 画图
plt.figure(figsize=(10, 6))
plt.boxplot(box_data, positions=weights, widths=10, patch_artist=True)
plt.plot(weights, mean_values, marker='o', linestyle='-', color='red', label='Mean')
# 标记均值
for i, mean_value in enumerate(mean_values):
    plt.text(weights[i], mean_value, f'{mean_value}', ha='center', va='bottom', color='black')

plt.title(title)
plt.xlabel(x_lable)
plt.ylabel(y_lable)
plt.legend()
plt.show()
```

```python

# 画图
plt.figure(figsize=(10, 6))
plt.plot(weights, mean_values, marker='o', linestyle='-', color='blue', label='Mean')
plt.fill_between(weights, np.array(mean_values) - np.array(std_values),
                 np.array(mean_values) + np.array(std_values), color='blue', alpha=0.2)

plt.title(title)
plt.xlabel(x_lable)
plt.ylabel(y_lable)
plt.legend()
plt.show()
```
